```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./kaggle_satelite_priv_class.csv", index_col='Class')

# originally each 0 - 0.1
df = 10 * df

df.columns = [
    '(1st) Kyle',
    '(2nd) n0ZF1Tuzrb3o',
    '(3rd) Vladimir & Sergey',
    '(4th) deepsense.io',
    '(5th) alno & kostia & deepsystems']
```

```python
# rder is not changed
df.plot(kind='bar', figsize=(12, 6))
plt.title("Dstl Satellite Imagery Feature Detection at Kaggle")
plt.ylabel("per-class Intersection over Union")
```

```python
# z-score
df_t = df.transpose()
df_z = ((df_t - df_t.mean()) / df_t.std()).transpose()
```

```python
df_z.plot(kind='bar', figsize=(10, 6))
plt.title("Dstl Satellite Imagery Feature Detection at Kaggle")
plt.ylabel("z-score of IoT (vs other teams)")
plt.legend(loc=(1.02, 0.75))
```

```python
df_z.transpose().plot(kind='bar', figsize=(8, 6))
plt.legend(loc=(1.02, 0.45))
plt.title("Dstl Satellite Imagery Feature Detection at Kaggle")
plt.ylabel('z-score of IoT (vs other teams)')
```
